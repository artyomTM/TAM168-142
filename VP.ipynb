```
from PIL import Image
def printQ(number, QF, QFT):
    if QFT=='.jpg':
        img=Image.open(QF+str(number)+QFT)
        img.show()
    elif QFT=='.txt':
        with open(QF+str(number)+QFT, 'r') as FILE:
            Q=FILE.read()
            print(Q)
def getAns(number, AF):
    with open(AF+str(number)+'.txt', 'r') as FILE:
        A=FILE.read()
    return A.lower()
def printR(CAC, AC):
    print(f"The End\nTrue Ans-s: {CAC}/{AC}")
    r=int(input("play?[0-no, 1-yes]>>>"))
    return r
```
